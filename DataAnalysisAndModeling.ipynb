# Telco Churn Classification Project

##  Goal
This python code analyzes data of a fictional telco company that provided home phone and internet services to 7043 customers in California in Q3. The goal is build a classification model that could predict if a customer will churn or not. Below are the questions used to understand the data even more.
* What is the overall percentage of churn value?
* What is the count of male vs. female by churn value?
* What is the churn value count by contract type?
* Is there a relationship between churn value and monthly charges?
* What's the most popular payment method?
* Do more people enroll in device protection?
* How many  people  go paperless?
* What are the top 5 reasons why customers churn?
* What is the most correlated factor to churn value?

## Data
The initial dataset contains 7042 rows, each representing a customer, and 33 columns.

### Variables
#### Demographics
CustomerID: A unique ID that identifies each customer.<br>
Count: A value used in reporting/dashboarding to sum up the number of customers in a filtered set.<br>
Gender: The customer’s gender: Male, Female<br>
Age: The customer’s current age, in years, at the time the fiscal quarter ended.<br>
Senior Citizen: Indicates if the customer is 65 or older: Yes, No<br>
Married: Indicates if the customer is married: Yes, No<br>
Dependents: Indicates if the customer lives with any dependents: Yes, No. Dependents could be children, parents, grandparents, etc.<br>
Number of Dependents: Indicates the number of dependents that live with the customer.<br>

#### Location
CustomerID: A unique ID that identifies each customer.<br>
Count: A value used in reporting/dashboarding to sum up the number of customers in a filtered set.<br>
Country: The country of the customer’s primary residence.<br>
State: The state of the customer’s primary residence.<br>
City: The city of the customer’s primary residence.<br>
Zip Code: The zip code of the customer’s primary residence.<br>
Lat Long: The combined latitude and longitude of the customer’s primary residence.<br>
Latitude: The latitude of the customer’s primary residence.<br>
Longitude: The longitude of the customer’s primary residence.<br>

### Population
ID: A unique ID that identifies each row.<br>
Zip Code: The zip code of the customer’s primary residence.<br>
Population: A current population estimate for the entire Zip Code area.<br>
 
### Services
CustomerID: A unique ID that identifies each customer.<br>
Count: A value used in reporting/dashboarding to sum up the number of customers in a filtered set.<br>
Quarter: The fiscal quarter that the data has been derived from (e.g. Q3).<br>
Referred a Friend: Indicates if the customer has ever referred a friend or family member to this company: Yes, No<br>
Number of Referrals: Indicates the number of referrals to date that the customer has made.<br>
Tenure in Months: Indicates the total amount of months that the customer has been with the company by the end of the quarter specified above.<br>
Offer: Identifies the last marketing offer that the customer accepted, if applicable. Values include None, Offer A, Offer B, Offer C, Offer D, and Offer E.<br>
Phone Service: Indicates if the customer subscribes to home phone service with the company: Yes, No<br>
Avg Monthly Long Distance Charges: Indicates the customer’s average long distance charges, calculated to the end of the quarter specified above.<br>
Multiple Lines: Indicates if the customer subscribes to multiple telephone lines with the company: Yes, No<br>
Internet Service: Indicates if the customer subscribes to Internet service with the company: No, DSL, Fiber Optic, Cable.<br>
Avg Monthly GB Download: Indicates the customer’s average download volume in gigabytes, calculated to the end of the quarter specified above.<br>
Online Security: Indicates if the customer subscribes to an additional online security service provided by the company: Yes, No<br>
Online Backup: Indicates if the customer subscribes to an additional online backup service provided by the company: Yes, No<br>
Device Protection Plan: Indicates if the customer subscribes to an additional device protection plan for their Internet equipment provided by the company: Yes, No<br>
Premium Tech Support: Indicates if the customer subscribes to an additional technical support plan from the company with reduced wait times: Yes, No<br>
Streaming TV: Indicates if the customer uses their Internet service to stream television programing from a third party provider: Yes, No. The company does not charge an additional fee for this service.<br>
Streaming Movies: Indicates if the customer uses their Internet service to stream movies from a third party provider: Yes, No. The company does not charge an additional fee for this service.<br>
Streaming Music: Indicates if the customer uses their Internet service to stream music from a third party provider: Yes, No. The company does not charge an additional fee for this service.<br>
Unlimited Data: Indicates if the customer has paid an additional monthly fee to have unlimited data downloads/uploads: Yes, No<br>
Contract: Indicates the customer’s current contract type: Month-to-Month, One Year, Two Year.<br>
Paperless Billing: Indicates if the customer has chosen paperless billing: Yes, No<br>
Payment Method: Indicates how the customer pays their bill: Bank Withdrawal, Credit Card, Mailed Check<br>
Monthly Charge: Indicates the customer’s current total monthly charge for all their services from the company.<br>
Total Charges: Indicates the customer’s total charges, calculated to the end of the quarter specified above.<br>
Total Refunds: Indicates the customer’s total refunds, calculated to the end of the quarter specified above.<br>
Total Extra Data Charges: Indicates the customer’s total charges for extra data downloads above those specified in their plan, by the end of the quarter specified above.<br>
Total Long Distance Charges: Indicates the customer’s total charges for long distance above those specified in their plan, by the end of the quarter specified above.<br>

### Status
CustomerID: A unique ID that identifies each customer.<br>
Count: A value used in reporting/dashboarding to sum up the number of customers in a filtered set.<br>
Quarter: The fiscal quarter that the data has been derived from (e.g. Q3).<br>
Satisfaction Score: A customer’s overall satisfaction rating of the company from 1 (Very Unsatisfied) to 5 (Very Satisfied).<br>
Satisfaction Score Label: Indicates the text version of the score (1-5) as a text string.<br>
Customer Status: Indicates the status of the customer at the end of the quarter: Churned, Stayed, or Joined<br>
Churn Label: Yes = the customer left the company this quarter. No = the customer remained with the company. Directly related to Churn Value.<br>
Churn Value: 1 = the customer left the company this quarter. 0 = the customer remained with the company. Directly related to Churn Label.<br>
Churn Score: A value from 0-100 that is calculated using the predictive tool IBM SPSS Modeler. The model incorporates multiple factors known to cause churn. The higher the score, the more likely the customer will churn.<br>
Churn Score Category: A calculation that assigns a Churn Score to one of the following categories: 0-10, 11-20, 21-30, 31-40, 41-50, 51-60, 61-70, 71-80, 81-90, and 91-100<br>
CLTV: Customer Lifetime Value. A predicted CLTV is calculated using corporate formulas and existing data. The higher the value, the more valuable the customer. High value customers should be monitored for churn.<br>
CLTV Category: A calculation that assigns a CLTV value to one of the following categories: 2000-2500, 2501-3000, 3001-3500, 3501-4000, 4001-4500, 4501-5000, 5001-5500, 5501-6000, 6001-6500, and 6501-7000.<br>
Churn Category: A high-level category for the customer’s reason for churning: Attitude, Competitor, Dissatisfaction, Other, Price. When they leave the company, all customers are asked about their reasons for leaving. Directly related to Churn Reason.<br>
Churn Reason: A customer’s specific reason for leaving the company. Directly related to Churn Category.<br>

Source: https://community.ibm.com/community/user/businessanalytics/blogs/steven-macko/2019/07/11/telco-customer-churn-1113

## Data Cleaning/Preprocessing

**Drop redundant columns**

CustomerID: Not necessary

Count: We can count the rows without it

Country: All USA

State: All California

Zip Code , Lat Long, Latitude, Longitude: One feature for location is enough

Churn label: We have the churn value

Churn Reason: Only a few people churned, meaning only they have a churn reason

```python
df = df.drop(['CustomerID',
              'Count', 
              'Country', 
              'State', 
              'Zip Code',
              'Lat Long', 
              'Latitude', 
              'Longitude', 
              'Churn Label',
              'Churn Reason'], axis=1)
```

```python
# Some total charges are not numbers
df['Total Charges'].value_counts()
```

```python
# Display lines where total charges are not numbers
not_numbers_mask = df['Total Charges'].str.contains(r'(\d+(?:\.\d+)?)', regex=True) == False
df[not_numbers_mask]
```

```python
df.loc[not_numbers_mask]['Tenure Months']
```

```python
df.loc[not_numbers_mask]['Monthly Charges']
```

```python
# Assuming those users are new customers (not a full month), impute the rows with half of monthly charges
# Assuming those users are new customers (not a full month), impute the rows with a charge of zero
df.loc[not_numbers_mask, 'Total Charges'] = 0

# Convert to float
df['Total Charges'] = df['Total Charges'].astype('float')
```

```python
df.loc[not_numbers_mask]['Total Charges']
```

```python
# Use Frequency Encoding for City
freq_map = df['City'].value_counts()
df['City'] = df['City'].map(freq_map)
df['City']
```

```python
# Get remaining string columns
is_string = (df.map(type) == str).all()
string_columns = list(is_string[is_string].index)
string_columns
```

```python
# Get columns with more than 2 unique values
nb_unique = df[string_columns].nunique()
more_than_two = nb_unique[nb_unique > 2]
more_than_two
```

```python
# Print the unique values
for column in more_than_two.index:
    print(column + ': ' + ', '.join(df[column].unique()))
```

```python
# Regroup data of Multiple Lines
df['Multiple Lines'] = np.where(df['Multiple Lines'] == 'No phone service', 'No', df['Multiple Lines'])
df['Multiple Lines'].unique()
```

```python
# Regroup data of Online Security
df['Online Security'] = np.where(df['Online Security'] == 'No internet service', 'No', df['Online Security'])
df['Online Security'].unique()
```

```python
# Regroup data of Online Backup
df['Online Backup'] = np.where(df['Online Backup'] == 'No internet service', 'No', df['Online Backup'])
df['Online Backup'].unique()
```

```python
# Regroup data of Device Protection
df['Device Protection'] = np.where(df['Device Protection'] == 'No internet service', 'No', df['Device Protection'])
df['Device Protection'].unique()
```

```python
# Regroup data of Tech Support
df['Tech Support'] = np.where(df['Tech Support'] == 'No internet service', 'No', df['Tech Support'])
df['Tech Support'].unique()
```

```python
# Regroup data of Streaming TV
df['Streaming TV'] = np.where(df['Streaming TV'] == 'No internet service', 'No', df['Streaming TV'])
df['Streaming TV'].unique()
```

```python
# Regroup data of Streaming Movies
df['Streaming Movies'] = np.where(df['Streaming Movies'] == 'No internet service', 'No', df['Streaming Movies'])
df['Streaming Movies'].unique()
```

```python
# Une One Hot Encoding for non-numeric columns

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded_cols = encoder.fit_transform(df[string_columns])
new_df = pd.DataFrame(encoded_cols, columns=encoder.get_feature_names_out(string_columns))
df = df.drop(columns=string_columns, axis=1).join(new_df)
```

```python
# Assert all columns are numeric
len(df.columns) == len(df.select_dtypes([np.number]).columns)
```

## Data Modeling

### Data Prep

```python
# Seperate values
X = df.drop('Churn Value', axis=1)
y = df['Churn Value']
```

```python
# train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
#We decided to keep 10% of the data for testing, since the volume of data is large
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

### Logistic Regression

```python
#imports 
from sklearn.linear_model import LogisticRegression
```

```python
# create model and fit
log_model=LogisticRegression() 
log_model.fit(scaled_X_train, y_train)
```

```python
#prediction
log_pred=log_model.predict(scaled_X_test)
log_pred
```

```python
log_cm = confusion_matrix(y_test, log_pred)
log_cm
```

```python
plt.rcParams['axes.grid'] = False
log_disp=ConfusionMatrixDisplay(log_cm, display_labels=log_model.classes_)
log_disp.plot()
plt.show()
```

```python
log_report = classification_report(y_test, log_pred)
print(log_report)
```

```python
#Interpretation
```

```python
log_model.intercept_
```

```python
log_model.coef_
```

### KNN

```python
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Initialize classifier with k=3
knn = KNeighborsClassifier(n_neighbors=3)

# Fit model
knn.fit(scaled_X_train, y_train)
```

```python
# Grid search
# Set-up pipeline object
pipe = Pipeline([
    ('scaler', StandardScaler()), 
    ('classifier', KNeighborsClassifier())
])
```

```python
param_grid = {'classifier__n_neighbors': list(range(1,20)),
              'classifier__weights': ['uniform', 'distance'],
              'classifier__p': np.linspace(1,2,5)}

full_cv_classifier = GridSearchCV(pipe,param_grid,cv=5,scoring='accuracy')

full_cv_classifier.fit(scaled_X_train,y_train)
```

```python
# View best params
full_cv_classifier.best_estimator_.get_params()
```

```python
pipe.set_params(classifier__n_neighbors=19)
```

```python
pipe.fit(scaled_X_train,y_train)
```

```python
pipe_pred = pipe.predict(scaled_X_test)
```

```python
# Classification report
print(classification_report(y_test,pipe_pred))
```

```python
# Confusion matrix
confusion_matrix(y_test,pipe_pred)
```

```python
cm = confusion_matrix(y_test, pipe_pred)
disp = ConfusionMatrixDisplay(cm, display_labels=pipe.classes_)
disp.plot();
```

### SVM

```python
from sklearn.svm import SVC
# The EDA showed the classes were unbalanced
# Add weights to alleviate the issue
class_weight = {0:3, 1:1}
svm = SVC(class_weight=class_weight)
```

```python
param_grid = {
    'C': [0.001, 0.01, 0.1, 1],
    'kernel': ['linear', 'rbf', 'poly'],
    'degree': [1, 2],
    'gamma': ['scale', 'auto']
}

grid = GridSearchCV(svm, param_grid)
grid.fit(scaled_X_train, y_train)
```

```python
# Best accuracy score
grid.best_score_
```

```python
# Best parameters
grid.best_params_
```

```python
svm_pred = grid.predict(scaled_X_test)

cm = confusion_matrix(y_test, svm_pred)
disp = ConfusionMatrixDisplay(cm, display_labels=grid.classes_)
disp.plot();
```

```python
report = classification_report(y_test, svm_pred)
print(report)
```

### Final Model

For our final model, we decided to go for the log model, because if we compare the 3 classification reports, this model hasd the better overall score and a better consistency. For the KNN model, its predictions scores were lower than log. For SVM, it had many different scores, which makes it not very reliable when predicting.

```python
#create and fit final model
final_model=LogisticRegression() 
final_model.fit(scaled_X_train, y_train)
```

```python
#prediction
final_predict=final_model.predict(scaled_X_test)
```

```python
#Display
final_cm = confusion_matrix(y_test, final_predict)
disp = ConfusionMatrixDisplay(final_cm, display_labels=grid.classes_)
disp.plot();
```
